```python
import pandas as pd
```

```python
df = pd.read_csv('russia_201901_1_tweets_csv_unhashed.csv', converters={"hashtags": lambda x: x.strip("[]").split(", ")},keep_default_na= False )
```

```python
df['size'] = [len(x) for x in df.hashtags]
```

```python
more = df[df['size']>1]
```
