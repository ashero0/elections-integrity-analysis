```python
# IRAN

num_sample_records = 10000

places = api.geo_search(query="Iran", granularity="country")
place_id = places[0].id

iran_id = []  # tweetid
iran_user_id = []  # userid
iran_user_name = []  # user_display_name
iran_user_screen_name = [] # user_screen_name
iran_user_rep_loc = []  # user_reported_location
iran_user_prof_desc = []  # user_profile_description
iran_user_profile_url = []  # user_profile_url
iran_follower_count = []  # follower_count
iran_following_count = []  # following_count
iran_account_creation_date = []  # account_creation_date
iran_account_language = []  # account_language
iran_tweet_text = []  # tweet_text
iran_tweet_time = []  # tweet_time
iran_tweet_client_name = []  # tweet_client_name
iran_in_reply_to_tweetid = []  # in_reply_to_tweetid
iran_in_reply_to_userid = []  # in_reply_to_userid
iran_is_retweet = []  # is_retweet
iran_geo = []  # geo
iran_like_count = []  # like_count
iran_retweet_count = []  # retweet_count
iran_hashtags = []  # hashtags
iran_urls = []  # urls
iran_user_mentions = []  # user_mentions

count = 0
for tweet in tweepy.Cursor(api.search, q = "place:%s" % place_id, count = num_sample_records, tweet_mode = 'extended').items(num_sample_records):
    iran_id.append(tweet.id)  # tweetid
    iran_user_id.append(tweet.user.id)  # userid
    iran_user_name.append(tweet.user.name)  # user_display_name
    iran_user_screen_name.append(tweet.user.screen_name)  # user_screen_name
    iran_user_rep_loc.append(tweet.user.location)   # user_reported_location
    iran_user_prof_desc.append(tweet.user.description)  # user_profile_description
    iran_user_profile_url.append(tweet.user.url)  # user_profile_url
    iran_follower_count.append(tweet.user.followers_count)  # follower_count
    iran_following_count.append(tweet.user.friends_count)  # following_count
    iran_account_creation_date.append(tweet.user.created_at)  # account_creation_date
    iran_account_language.append(tweet.user.lang)  # account_language
    iran_tweet_text.append(tweet.full_text)  # tweet_text
    iran_tweet_time.append(tweet.created_at)  # tweet_time
    iran_tweet_client_name.append(tweet.source)  # tweet_client_name
    iran_in_reply_to_tweetid.append(tweet.in_reply_to_status_id)  # in_reply_to_tweetid
    iran_in_reply_to_userid.append(tweet.in_reply_to_user_id)  # in_reply_to_userid
    iran_geo.append(tweet.geo)  # geo
    iran_like_count.append(tweet.favorite_count)  # like_count
    iran_retweet_count.append(tweet.retweet_count)  # retweet_count
    iran_hashtags.append(tweet.entities["hashtags"])  # hashtags
    iran_urls.append(tweet.entities["urls"])  # urls
    iran_user_mentions.append(tweet.entities["user_mentions"])  # user_mentions
    
    if (hasattr(tweet, 'retweeted_status')):
        iran_is_retweet.append(True)
    else:
        iran_is_retweet.append(False)
    
    if (count % 1000 == 0):
        print(count)
    count = count + 1
    
```

```python
iran_dict = {
    'tweetid': iran_id,
    'userid': iran_user_id,
    'user_display_name': iran_user_name,
    'user_screen_name': iran_user_screen_name,
    'user_reported_location': iran_user_rep_loc,
    'user_profile_description': iran_user_prof_desc,
    'user_profile_url': iran_user_profile_url,
    'follower_count': iran_follower_count,
    'following_count': iran_following_count,
    'account_creation_date': iran_account_creation_date,
    'account_language': iran_account_language,
    'tweet_text': iran_tweet_text,
    'tweet_time': iran_tweet_time,
    'tweet_client_name': iran_tweet_client_name,
    'in_reply_to_tweetid': iran_in_reply_to_tweetid,
    'in_reply_to_userid': iran_in_reply_to_userid,
    'is_retweet': iran_is_retweet,
    'geo': iran_geo,
    'like_count': iran_like_count,
    'retweet_count': iran_retweet_count,
    'hashtags': iran_hashtags,
    'urls': iran_urls,
    'user_mentions': iran_user_mentions,
}

iran = pd.DataFrame(iran_dict)
# print(iran['tweet_text'])
# print(iran['is_retweet'])
```

```python
iran.to_csv('iran_baseline_tweets.csv')
```
