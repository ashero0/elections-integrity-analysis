```python
import pandas as pd
```

```python
from googletrans import Translator
translator = Translator()
```

```python
import emoji

def remove_emoji(text):
    return emoji.get_emoji_regexp().sub(u'', text)
```

```python
iran= pd.read_csv('samples/iran_sample.csv', delimiter=',', error_bad_lines=False)
russia= pd.read_csv('samples/rus_sample.csv', delimiter=',', error_bad_lines=False)
venez= pd.read_csv('samples/ven_sample.csv', delimiter=',', error_bad_lines=False)
```

```python
#translation = translator.translate('안녕하세요.')
#print(translation.text)

data_text_Iran = iran[['tweet_text']]
for tweet in data_text_Iran['tweet_text']:
   
    tweet2 = remove_emoji(tweet)
    print(tweet2)
   
    translation = translator.translate(tweet2)
    print(translation.text, "\n\n")
    tweet = translation.text
data_text_Iran['index'] = data_text_Iran.index
documents_Iran = data_text_Iran
```

```python
data_text_Russia = russia[['tweet_text']]
for tweet in data_text_Russia['tweet_text']:
    tweet = remove_emoji(tweet)
    translation = translator.translate(tweet)
    tweet = translation.text
data_text_Russia['index'] = data_text_Russia.index
documents_Russia = data_text_Russia
```

```python
data_text_Venez = venez[['tweet_text']]
for tweet in data_text_Venez['tweet_text']:
    tweet = remove_emoji(tweet)
    translation = translator.translate(tweet)
    tweet = translation.text
data_text_Venez['index'] = data_text_Venez.index
documents_Venez = data_text_Venez
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
np.random.seed(2018)
import nltk
from nltk.stem import PorterStemmer as ps
nltk.download('wordnet')
```

```python
def lemmatize_stemming(text):
    stemmer = ps()
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

def preprocess(text):
    
    
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and 'http' not in token and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

```python
documents_Iran = documents_Iran.dropna(subset=['tweet_text'])
documents_Russia = documents_Russia.dropna(subset=['tweet_text'])
documents_Venez = documents_Venez.dropna(subset=['tweet_text'])

```

```python

processed_docs_Iran = documents_Iran['tweet_text'].map(preprocess)
processed_docs_Russia = documents_Russia['tweet_text'].map(preprocess)
processed_docs_Venez = documents_Venez['tweet_text'].map(preprocess)
```

```python

dictionary_Iran = gensim.corpora.Dictionary(processed_docs_Iran)
dictionary_Russia = gensim.corpora.Dictionary(processed_docs_Russia)
dictionary_Venez = gensim.corpora.Dictionary(processed_docs_Venez)

count = 0
for k, v in dictionary_Iran.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break

count = 0
for k, v in dictionary_Russia.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
        
count = 0
for k, v in dictionary_Venez.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
```

```python
dictionary_Iran.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
dictionary_Russia.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
dictionary_Venez.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
```

```python
bow_corpus_Iran = [dictionary_Iran.doc2bow(doc) for doc in processed_docs_Iran]
bow_corpus_Russia = [dictionary_Russia.doc2bow(doc) for doc in processed_docs_Russia]
bow_corpus_Venez = [dictionary_Venez.doc2bow(doc) for doc in processed_docs_Venez]
```

```python
from gensim import corpora, models
tfidf_Iran = models.TfidfModel(bow_corpus_Iran)
corpus_tfidf_Iran = tfidf_Iran[bow_corpus_Iran]

tfidf_Russia = models.TfidfModel(bow_corpus_Russia)
corpus_tfidf_Russia = tfidf_Iran[bow_corpus_Russia]

tfidf_Venez = models.TfidfModel(bow_corpus_Venez)
corpus_tfidf_Venez = tfidf_Iran[bow_corpus_Venez]

from pprint import pprint
```

```python
lda_model_Iran = gensim.models.LdaMulticore(bow_corpus_Iran, num_topics=20, id2word=dictionary_Iran, passes=2, workers=2)
lda_model_Russia = gensim.models.LdaMulticore(bow_corpus_Russia, num_topics=20, id2word=dictionary_Russia, passes=2, workers=2)
lda_model_Venez = gensim.models.LdaMulticore(bow_corpus_Venez, num_topics=20, id2word=dictionary_Venez, passes=2, workers=2)
```

```python
for idx, topic in lda_model_Iran.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```
